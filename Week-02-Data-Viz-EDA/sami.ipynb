# Make a timeline line chart in which the x-axis is the Year, and the y-axis is the mean MSRP.
* you're going to have to do a `groupby` for this one for sure. 
* What's noticeable about it and what do you think the error is...


```python
# Make a timeline line chart in which the x-axis is the year, and the y-axis is the average MSRP.
# make_count_s_time_line = make_count_s.groupby('Year')

# # Check data types
# print(make_count.dtypes)

# # Reset the index if it's part of a multi-level index
# make_count.reset_index(inplace=True)

average_MSRP = df.groupby("Year")['MSRP'].mean()

plt.figure(figsize= (16, 12))

plt.plot(average_MSRP.index, average_MSRP.values, marker ='o', linestyle= '-')

plt.xlabel('Year')
plt.ylabel('Average MSRP')
plt.title('Average MSRP over time')

# average_MSRP.plot(kind= "line chart", x= 'Year', y = 'MSRP')
```

# It seems as though in the years before (and includig) 2000, they were counting in tens.
Make a new column that is called `adjusted_price`, that contains all prices, however, for every year before 2000 make it 10x the original MSRP.  
   * Hint; you might need to use our old friend `np.where`

```python
# Make a column where is 10 when year is less than 2000, else 1.
condition = df['Year']<=2000
df["adjusted_price"] = np.where(condition, 10, 1)


# Multiply the MSRP by the price adjuster.

df["adjusted_price"] = df["MSRP"]* df["adjusted_price"] 
df.head(100)
```

# Replot the new adjusted price.  
* Make the y-axis start at 0 and go up to 100,000

```python
# Plot new prices

adjusted_price_average_MSRP = df.groupby("Year")['adjusted_price'].mean()

plt.figure(figsize= (16, 12))

plt.plot(adjusted_price_average_MSRP.index, adjusted_price_average_MSRP.values, marker ='o', linestyle= '-')

plt.xlabel('Year')
plt.ylabel('Adjusted Price Average MSRP')
plt.title('Average MSRP over time')

# Making the y-axis start at 0 and go up to 100,000

plt.ylim(0, 100000)
```

# Find which features actually have the strongest linear relationship using correlations.
* Make a heatmap plot of all of the correlations in our dataset.
* Change the figure size of our heatmap plot to be 8x8
* __Which feature does Engine HP have the strongest relationship with, and why do you think that relationship exists.__

```python
# * Make a heatmap plot of all of the correlations in our dataset.
# * Change the figure size of our heatmap plot to be 8x8

# setting plot size
plt.figure(figsize=(8, 8))

sns.heatmap(df.corr(), 
            annot=True, 
            cmap='coolwarm');

# Show the heatmap
plt.show()
```

```python
# Which feature does Engine HP have the strongest relationship with, and why do you think that relationship exists.

print('From the heatmap I can say Engine HP has strongest relationship with Engine Cylinders.\nThe relationship between engine horsepower and engine cylinders exists because the more cylinders an engine has,\nthe more air and fuel it can burn per combustion cycle. This is because each cylinder has its own intake and exhaust valves, which allows for more efficient combustion. As a result, more cylinders can produce more power.')
```
